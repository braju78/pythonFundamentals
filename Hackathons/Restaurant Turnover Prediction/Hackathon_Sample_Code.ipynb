```python
## Importing the libraries

import pandas as pd
from sklearn.linear_model import LinearRegression
```

# Loading the Training Data and using the Training Data to build the Model

```python

df_train = pd.read_csv('/content/Train_dataset.csv')
data = df_train.copy()
df_train.head()
```

```python
## Checking for the null values

df_train.isnull().sum()
```

```python
## Importing Simple Imputer to treat the null values

from sklearn.impute import SimpleImputer
```

```python
## Defning the Simple Imputer funtion to use 'mean' as a strategy of imputation

SI = SimpleImputer(strategy='mean', copy=False)
```

```python
## Fitting the Simple Imputer function to get the mean of the variable

SI.fit(df_train[['Instagram Popularity Quotient']])
```

```python
## Transforming the variable --> Imputing the variable with its mean

df_train[['Instagram Popularity Quotient']] = SI.transform(df_train[['Instagram Popularity Quotient']])
```

```python
SI.fit(df_train[['Facebook Popularity Quotient']])
df_train[['Facebook Popularity Quotient']] = SI.transform(df_train[['Facebook Popularity Quotient']])
```

```python
SI.fit(df_train[['Resturant Tier']])
df_train[['Resturant Tier']] = SI.transform(df_train[['Resturant Tier']])

SI.fit(df_train[['Overall Restaurant Rating']])
df_train[['Overall Restaurant Rating']] = SI.transform(df_train[['Overall Restaurant Rating']])


SI.fit(df_train[['Live Music Rating']])
df_train[['Live Music Rating']] = SI.transform(df_train[['Live Music Rating']])
```

```python
SI.fit(df_train[['Comedy Gigs Rating']])
df_train[['Comedy Gigs Rating']] = SI.transform(df_train[['Comedy Gigs Rating']])

SI.fit(df_train[['Value Deals Rating']])
df_train[['Value Deals Rating']] = SI.transform(df_train[['Value Deals Rating']])

SI.fit(df_train[['Live Sports Rating']])
df_train[['Live Sports Rating']] = SI.transform(df_train[['Live Sports Rating']])

SI.fit(df_train[['Ambience']])
df_train[['Ambience']] = SI.transform(df_train[['Ambience']])
```

```python
df_train.describe().T
```

```python
# prompt: Get all the non-numeric columns from df_train

non_numeric_cols = df_train.select_dtypes(exclude=['number']).columns
print(non_numeric_cols)

df_train = pd.get_dummies(df_train, columns=non_numeric_cols)

df_train.describe().T
```

```python
df_train.info()
```

```python
## Data preparation for model building
X = df_train.drop('Annual Turnover',axis=1)
y = df_train['Annual Turnover']
```

```python
# to split the data into train and test sets
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(
    X,    # specifying the independent variables
    y,    # specifying the dependent variable
    test_size=0.20,    # specifying the size of the test set as a fraction of the whole data
    random_state=42    # specifying a seed value to enable reproducible results
)
```

```python
## Declaring the Linear Regression function

LR = LinearRegression()
```

```python
## Fitting the Linear Regression function

model = LR.fit(X_train,y_train)
```

```python
## Checking the score of the function on the training data

model.score(X_train,y_train)
```

# Loading the Test Data and using the Test Data to Predict the target Variable

```python
df_test = pd.read_csv('/content/Test_dataset.csv')
df_test.head()
```

```python
## Checking for the null values

df_test.isnull().sum()
```

```python
## Using the model built on the Training set to predict on the Test Set

prediction = model.predict(df_test[['Instagram Popularity Quotient']])
prediction
```
